```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Dropout ve EarlyStopping' i overfitting' e karşı kullanacağız.
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))    # Önerilen giriş nöron sayısı, veri setindeki kolon yani öznitelik sayısıdır.
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))  # Çıkış katmanında aktivasyon fonksiyonuna sigmoid yazılması 0-1 arasında değer elde etmemize yarar.

model.compile(loss="binary_crossentropy",optimizer = "adam")

model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot()
```

**Yukarıdaki grafikte overfitting olduğu görülüyor. Biz modele bu uçurum arttığında epochu durdur komutu veriririz. Bunu da `EarlyStopping` ile yaparız.**




```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")

earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)  
# val_loss u takip et. val_loss u min. tutmaya çalış. Bozulmaya 25 epoch sabret, 25 epoch içinde iyileşme olmuyorsa durdur.
# val_loss tan başka şey de takip edebilirdik. Buna bağlı olarak min max modu da değişebilirdi.

model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot()
```

**Droupout: Katmanlardan belli oranda veriyi rasgele atarak overfitting engellemeye çalışır.**

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")

model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
kayipDf.plot()
```

```python
tahminlerimiz = (model.predict(x_test) > 0.5).astype("int32")
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,tahminlerimiz))
```

```python
print(confusion_matrix(y_test,tahminlerimiz))
```
