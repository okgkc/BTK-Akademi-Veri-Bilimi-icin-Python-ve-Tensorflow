## Model Oluşturma

```python
y = dataFrame["price"].values
x = dataFrame.drop("price",axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train), len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")

model.fit(x=x_train, y = y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)  

# validation_data=(x_test,y_test) değeri manuel olarak yaptığımız kıyaslama işini otomatikleştiriyor
# batch_size=250 veri seti büyük olduğu için verileri 250lik partiler olarak verdik.
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

**Eğer grafikte bir noktadan sonra loss ve val_loss grafikleri ayrılıyorsa aşırı öğrenme başlamıştır. Epoch değerinin düşürülmesi gerekir.**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
tahminDizisi = model.predict(x_test)
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
